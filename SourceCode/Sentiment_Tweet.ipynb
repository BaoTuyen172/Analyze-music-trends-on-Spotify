```python
import nltk
nltk.download('vader_lexicon')

import numpy as np
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
twitter_stream = twitter.TwitterStream(auth=auth)
iterator = twitter_stream.statuses.sample()
analyzer = SentimentIntensityAnalyzer()
```

```python
import csv
score=[]
datatwiter=[]
for tweet in tweets:
  score = analyzer.polarity_scores(tweet)
  print(score)
  datatwiter.append([tweet,score])
```

```python
print(datatwiter)
```

```python
import csv
with open('/content/drive/My Drive/Colab Notebooks/Data/SomeThingJustLikeThis.csv', mode='w', newline='') as csv_file:
  obj=csv.writer(csv_file)
  fieldnames = ['tweet', 'score']
  obj.writerow(fieldnames) 
  for row in datatwiter:
    obj.writerow(row)
  csv_file.close()
print("Lưu thành công")
```
